The goal of VAE is to generate a realistic image given a random vector that is generated from a pre-defined distribution.

The stategy is as follows:

1. Encoder network: Given an image, the encoder network will generate two vectors: mean and variance. The mean and variance will be used to generate a random vector from a pre-defined distribution. We sum the mean and variance and multiply it by a random vector from a pre-defined distribution. The result is the input to the decoder network.

2. Decoder network: Given the input from the encoder network, the decoder network will generate an image. The loss function is combination of L2 distance and KL divergence. The L2 distance is the difference between the generated image and the original image. The KL divergence is the difference between the distribution of the random vector and the pre-defined distribution. 


The mean vector has distribution around 0 and the variance vector has distribution around 1. The pre-defined distribution is a normal distribution with mean 0 and variance 1. The KL divergence is calculated as follows:

$$KL = - \sum \frac{1}{2} (1 + \log(\sigma^2) - \mu^2 - \sigma^2)$$

where p(x) is the distribution of the random vector and q(x) is the pre-defined distribution.

The loss function is calculated as follows:

$$L = \sum_{i=1}^{n} (x_i - \hat{x_i})^2 + KL$$

where x is the original image and $\hat{x}$ is the generated image.

### VAE Class

VAE class with 2 encoding layers + 1 latent layer + 2 decoding layers

```python
#  build a VAE ( Variational Autoencoder) with latent dimension = 5 


class VAE2:

    def __init__(self, input_dim, latent_dim):
        self.input_dim = input_dim
        self.latent_dim = latent_dim
        self.encoder = None
        self.decoder = None
        self.model = None
        
        self._build()

    def sampling(self, args):
        z_mean, z_log_var = args
        batch = K.shape(z_mean)[0]
        dim = K.int_shape(z_mean)[1]
        epsilon = K.random_normal(shape=(batch, dim))
        return z_mean + K.exp(0.5 * z_log_var) * epsilon

    def _build(self):
        # encoder
        inputs = Input(shape=(self.input_dim, self.input_dim, 1), name='encoder_input')
        x = inputs

        # 64 filters of 3x3 each with stride 2 and relu activation function
        for i in range(2):
            filters = 32 * 2 ** (2 - i)
            x = Conv2D(filters=filters,
                       kernel_size=3,
                       activation='relu',
                       strides=2,
                       padding='same')(x)

        # shape info needed to build decoder model
        shape = K.int_shape(x)

        # generate latent vector Q(z|X)
        x = Flatten()(x)

        # mu and sigma for latent vector z
        self.mu = Dense(self.latent_dim, name='latent_mu')(x)
        self.log_var = Dense(self.latent_dim, name='latent_log_var')(x)

        # use reparameterization trick to push the sampling out as input
        z = Lambda(self.sampling, output_shape=(self.latent_dim,), name='z')([self.mu, self.log_var])

        # instantiate encoder model
        self.encoder = Model(inputs, [self.mu, self.log_var, z], name='encoder')
        self.encoder.summary()

        # build decoder model

        # latent inputs to reconstruct inputs from latent space
        latent_inputs = Input(shape=(self.latent_dim,), name='z_sampling')

        # reshape to (7, 7, 32) which is the shape of the last conv layer in encoder
        x = Dense(shape[1] * shape[2] * shape[3], activation='relu')(latent_inputs)

        # reshape to (7, 7, 32)
        x = Reshape((shape[1], shape[2], shape[3]))(x)

        # use Conv2DTranspose to reverse the conv layers from the encoder
        for i in range(2):
            filters = 32 * 2 ** i
            x = Conv2DTranspose(filters=filters,
                                kernel_size=3,
                                activation='relu',
                                strides=2,
                                padding='same')(x)

        outputs = Conv2DTranspose(filters=1,
                                  kernel_size=3,
                                  activation='sigmoid',
                                  padding='same',
                                  name='decoder_output')(x)

        # instantiate decoder model
        self.decoder = Model(latent_inputs, outputs, name='decoder')
        self.decoder.summary()

        # instantiate VAE model
        outputs = self.decoder(self.encoder(inputs)[2])
        self.model = Model(inputs, outputs, name='vae_mlp')

        reconstruction_loss = mse(K.flatten(inputs), K.flatten(outputs))
        reconstruction_loss *= self.input_dim * self.input_dim
        kl_loss = 1 + self.log_var - K.square(self.mu) - K.exp(self.log_var)
        kl_loss = K.sum(kl_loss, axis=-1)
        kl_loss *= -0.5
        vae_loss = K.mean(reconstruction_loss + kl_loss)
        self.model.add_loss(vae_loss)
        self.model.compile(optimizer='adam')
        self.model.summary()
        

    
```
